* <p style = 'font-size:20px'> Nesta aula, vamos utilizar um conjunto de dados acadêmicos para realizarmos um teste estatístico.</p>

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Importando os dados .

df = pd.read_csv('grades.csv')
df.head(3)
```

```python
# Convertendo para datetime.
df['assignment1_submission'] = pd.to_datetime(df['assignment1_submission'])
df['assignment1_submission'].head()
```

```python
# Vamos criar duas categorias de alunos: aquela com os que enviaram a primeira tarefa em 2015, e outra com
# os que enviaram depois.
import datetime
def date_sep(index):
    if index <= datetime.datetime(2016,1,1):
        return '2015 Submissions'
    else:
        return 'Later Submissions'
    
submissions = (df.set_index('assignment1_submission')
    .groupby(by = date_sep)['assignment1_grade'])
```

* <p style = 'font-size:20px'> As notas dos projetos enviados em 2015 são iguais àquelas dos trabalhos enviados depois?</p>

* <p style = 'font-size:20px'> Para respondermos isso, vamos realizar um teste t-Student. Vamos definir a hipótese nula como "as notas são iguais" e a alternativa "as notas NÃO são iguais". Nosso nível de significância será de 5%</p>

```python
from scipy.stats import ttest_ind
submissions_2015, submissions_later = submissions

# Realizando o teste t-Student.
ttest_ind(submissions_2015[1], submissions_later[1])

# Como o p-valor é maior do que o nível de significância, não podemos rejeitar a hipótese nula!.
```

* <p style = 'font-size:20px'> Vamos fazer esse mesmo teste para dois DataFrames com números aleatórios.</p>

```python
# Cada DF terá cem linhas e cem colunas.
df1 = pd.DataFrame(np.random.rand(100,100))
df1.head()
```

```python
df2 = pd.DataFrame(np.random.rand(100,100))
df2.head()
```

```python
# Vamos utilizar o método 'ttest_ind' para cada coluna dos DF's. Será que elas são iguais?
# Vamos estabelecer o nível de significância como 10%.

def tstudent(alpha = 0.1):
    diff = 0
    for i in df1.columns:
        test = ttest_ind(df1[i], df2[i])
        if test[1] < alpha:
            print(f'The DataFrames are unequal in the column:{i}')
            diff+=1
    print(f'{diff / 100:.2%} das colunas são diferentes')
            
tstudent(alpha = 0.1)
```

```python
# Observe como a porcentagem dos casos de diferença se modela de acordo com o valor do nível de significância dado.
tstudent(alpha = 0.05)
```
