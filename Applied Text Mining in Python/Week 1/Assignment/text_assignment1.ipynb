```python
# Uma das datas numéricas está sendo considerada como data de string.
# O seu texto tem a palavra 'Marked', que cai como uma padrão cabível em 'month_pattern'...
import re
def not_year_two(l):    
    for i in range(len(l)):
        whole, day, month, year = l[i]
        if day == '' and month =='' and len(year)==2:
            continue
        else:
            return l[i]
        
def find_date(series):
    indices = []
    new_dates = []
    months_list=['Jan', 'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    month_pattern='(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z.]*'
    prohibited_terms = '(Marked|Marital|Declined)'
    # A dictionary with the month abbreviation with its respective ordinal number (não remover, por ora).
    month_dict={months_list[i]:i+1 for i in range(len(months_list))}
    for date in series:
        #teste
        index = series[series == date].index[0]
        #print(index)
        final_date=None
        date = re.sub(prohibited_terms,'', date)
        # If the date have a word representing the month, follow this logic.
        if re.search(month_pattern, date):
            alpha='((\d{1,2})? ?(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z.]* ?(\d{1,2})?[,]? (\d{4}))'
            # Stripping possible undesirable spaces.
            date_extracted = re.findall(alpha, date)[0][0].strip()
            date_extracted = date_extracted.replace('Janaury','January').replace('Decemeber','December').replace('2June', '2 June')
            #print(date_extracted)
            # If date has no day number, insert a 1.
            #print(date_extracted)
            if len(date_extracted.split()) == 2:
                date_extracted = '1 '+date_extracted.strip()
            final_date = date_extracted.replace(',','').replace('.','')
        # Else, the date must be numeric!
        else:
            #print(date)
            numeric='((\d{,2}[/-])?(\d{,2}[/-])?(\d{2,4}))'
            # Extracting the potential dates
            potential_dates = re.findall(numeric, date)
            date_extracted = not_year_two(potential_dates)[0].replace('/','-')
            # Splitting the date on the dash.
            date_splitted = date_extracted.split('-')
            year = int(date_splitted[-1])
            # The regex might catch some random numbers that are not a date in any way.
            if len(date_splitted) == 1 and len(str(year)) <4:
                continue
            elif len(date_splitted) == 1 and len(str(year)) ==4:
                date_splitted.insert(0,str(1))
                date_splitted.insert(0,str(1))
                final_date='-'.join(date_splitted)
                #print(final_date)
            # If the given date has only the month and year numbers, add '1' as day
            elif len(date_splitted) == 2:
                date_splitted.insert(0,str(1))
                final_date = '-'.join(date_splitted)
            else:
                final_date=date_extracted
                
        #print(date_extracted,pd.to_datetime(final_date).date())
        # teste
        indices.append(index)
        new_dates.append(pd.to_datetime(final_date).date())
    extra_dates = ['8/9/97','11/14/83', '4/2013', '2007']
    for dt in extra_dates:
        new_dates.append(pd.to_datetime(dt).date())
        
    values = pd.Series(new_dates)#.sort_values().index
    return pd.Series(values).sort_values()
                
find_date(series)
```
