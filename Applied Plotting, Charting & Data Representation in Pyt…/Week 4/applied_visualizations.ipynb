```python
# Vamos utilizar a matriz de scatter plots como um exemplo.
# Esse gráfico composto é capaz de evidenciar as relações entre os dados das colunas, além de montar histogramas.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('iris.csv')
df.head()
```

```python
# Fazendo a famosa scatter matrix.
pd.plotting.scatter_matrix(df, figsize = (8,5));
```

```python
# O parallel_coordinates nos ajuda a evidenciar diferenças de valores entre categorias distintas (class_column) 
# presentes no DF.
# No caso, estamos analisando diferenças de estatura entre diferentes espécies da flor iris.
pd.plotting.parallel_coordinates(df, class_column = 'Name')
```

<h1 style = 'font-size:40px'> Plotting with Seaborn</h1>

```python
# joinplot -> equivalente ao scatter plot com histogramas visto no módulo 2.
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0,10,1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60,15,1000), name='v2')
```

```python
import seaborn as sns
sns.jointplot(v1, v2);
```

```python
# O jointplot também possui um argumento 'kind', que nos permite criar diferentes estilos de gráfico central.
sns.jointplot(v1, v2, kind = 'hex');
```

```python
# Assim como o matplotlib, o seaborn também possui um conjuto de estilos de gráficos pé-definidos.
sns.set_style('white')
sns.jointplot(v1, v2, kind = 'kde', space =0);
```

```python
# O pairplot é capaz de criar scatter matrices com outros gráficos que não sejam histogramas.bb
df = pd.read_csv('iris.csv')
sns.pairplot(df, hue = 'Name')
```

```python
# Swarm plots e violin plots podem ser coniderados versões mais avançadas do boxplot.
plt.figure(figsize=(8,6))
plt.subplot(121)
sns.swarmplot('Name', 'PetalLength', data=df);
plt.subplot(122)
sns.violinplot('Name', 'PetalLength', data=df);
```
