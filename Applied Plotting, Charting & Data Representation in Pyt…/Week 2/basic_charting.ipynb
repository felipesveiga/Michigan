```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [1,2,3])
```

```python
# plt.gca -> nos dá acesso aos eixos do gráfico.
import numpy as np
x = np.linspace(-100, 100, 10001)
y = (x ** 2) + (2*x) +5
plt.plot(x,y)

# Vamos definir uma variável que se refira aos eixos do gráfico.
ax = plt.gca()

# Inserindo os limtes dos eixos.
ax.axis([-10, 10, -100, 100])

# Printando
ax.get_children()
```

<p style = 'font-size:40px'> <strong>Scatter Plots</strong> </p>

```python
# Os scatter plots são bastante úteis na evidenciação de relação entre duas variáveis.
x = [i for i in range(20)]
y = [i for i in range(0, 100, 5)]
plt.scatter(x, y, label = 'Points')

# Adicionando uma caixa de legenda.
plt.legend(loc = 7, frameon = False, title = 'Legend')
```

<p style = 'font-size:40px'> <strong>Line Plots</strong> </p>

```python
import numpy as np
y1 = np.linspace(0, 5, 11)
y2 = y1 ** 2

# Podemos ver que, mesmo não definindo um eixo x, o matplotlib foi capaz de deduzir que queríamos plotar dois gráficos
# distintos.
plt.plot(y1, '-o', y2, '-o')

# Aqui, usaremos a fill_between. Como não definimos um eixo x, vamos ter que fazer um artifício com 'range'.
plt.fill_between(range(len(y1)), y1, y2, alpha = 0.25)
```

```python
# Time Series com o matplotlib.
# A biblioteca também suporta dados de datas.
import pandas as pd
dates = pd.date_range(start = '2001-10-24', end = '2002-01-01', freq = 'W')
y = [i * r for i in range(5) for r in range(2)]
plt.figure(figsize = (15, 5))
plt.plot(dates, y)

# Vamos mudar a inclinação dos itens do eixo x.
x = plt.gca().xaxis

for item in x.get_ticklabels():
    item.set_rotation(45)
```

<p style = 'font-size:40px'> <strong>Bar Charts</strong> </p>

```python
y = [i for i in range(2, 11, 2)]
e = [i for i in range(5)]

# Os gráficos de barra podem conter também
plt.bar(range(len(y)) , y, width = 0.3, yerr = e)
```

```python
# Stacked bar charts usando o argumento 'bottom'.
a = [i for i in range(0, 20, 5)]
b = [i for i in range(4)]

plt.bar(range(len(a)),a)
plt.bar(range(len(b)),b, bottom = a)
```

```python
# Stacked horizontal bar charts com 'left'
a = [i for i in range(0, 20, 5)]
b = [i for i in range(4)]

plt.barh(range(len(a)),a)
plt.barh(range(len(b)),b, left = a)
```

```python
# DICA!
# COMO FORMATAR AS DATAS DO EIXO X
import matplotlib.dates as mdates
xfmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(xfmt)
```

<p style = 'font-size:40px'> <strong>Dejunkifying a Plot</strong> </p>

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

bars = plt.bar(pos, popularity, align='center')
plt.xticks(pos, labels = [])
plt.yticks(popularity, labels = [])
plt.ylabel('% Popularity')
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# Removendo a moldura do gráfico.
ax = plt.gca()
for x in ax.spines.values():
    x.set_visible(False)
#TODO: remove all the ticks (both axes), and tick labels on the Y axis

# Dica, podemos iterar sobre as barras de nosso gráfico e obter certos valores de cada uma delas.
for bar in bars:
    print(bar.get_height(), bar.get_width())

plt.show()
```
