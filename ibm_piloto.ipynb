```python
import numpy as np
import math
```

```python
# Os arrays do numpy
array = np.array([[4,5,6],[1,2,3]])

# Curiosidade: podemos utilizar ndim para verificar a quantidade de dimensões de um array do numpy
print(array.ndim)

# O Python também pode nos informar a natureza da matriz produzida com shape
print(array.shape)

# Ou seja, 'array' é uma matriz 2X3
```

```python
# Pegando o tipo de dado em nosso np.array
# Os elementos dos arrays do numpy devem ser todos da mesma espécie (só integers ou só floats)
array.dtype.name
```

```python
# Arrays especiais: o numpy oferece uma série de arrays pré-definidos para utilizarmos

# Array de 0's
zeros = np.zeros((2,3)) 
print(zeros)

# Arrays de 1's
uns = np.ones((4,4))
print(uns)

# Como podemos ver, o shape do array é passado como uma tupla dentro dos parênteses de argumentos.
# Lembre-se, são parênteses dentro de parênteses!

# Array's com números aleatórios [0,1[
aleatorios = np.random.rand(5,5)
print(aleatorios)

# No entanto, observe que para np.random.rand a história é diferente: não devemos passar uma tupla nos parênteses
                                                                                                    # de argumentos
```

```python
# Podemos também criar arrays baseados em um determinado range de números
# Vamos fazer uma Progressão Aritmética de razão 3 no intervalo [20, 60[
alcance = np.arange(20,60,3)
print(alcance)

# Poderíamos aproveitar esse array para calcularmos, por exemplo, a soma de seus números!
print(alcance.sum())
```

```python
# linspace: nos dará um array de floats. Start seria o valor mínimo do array; stop, o máximo
# Veja que o resultado será uma PA que se inicia no valor start e termina no stop
# O elemento num seria o número de elementos do array

np.linspace(start=1, stop=5, num=10)
```

## Operações entre arrays

```python
# Assim como na matemática, as matrizes do numpy podem ser objetos de cálculos

# Soma
a = np.array([[1,2,3],[0,8,6]])
b = np.random.rand(2,3)
e = '\n\n'
print(a+b, end=e)

# Subtração
print(a-b, end=e)

# Multiplicação
print(a*b, end=e)

# Divisão
print(a/b, end=e)
```

```python
# As operações com arrays do numpy podem ser úteis em muitas áreas de nossas vidas.
# Vamos pegar o peso de alguns jogadores de futebol
from pandas import read_csv
players = read_csv('players_fifa22.csv', usecols=['Name','Weight'], index_col=['Name'], nrows=20)

# No DF, as massas dos jogadores estão em kg.
# Vamos convertê-las a lbs
weights = np.array([weight for weight in players['Weight']])
lbs = weights * 2.204623 # O valor ao lado é a quantidade de lbs para um kg
print(lbs)
```

```python
# Boolean arrays: arrays com valores booleanos para uma data condição
# Digamos que nós quiséssemos saber quais jogadores de lbs possuem mais do que 185lbs
heavy = lbs>185
print(heavy)

# Operações matemáticas também são aceitas na construção de boolean arrays
um_a_dez = np.arange(1,11)
pares = um_a_dez % 2 == 0 
print(pares)
```

```python
# Multiplicação entre matrizes.
# Para que o numpy faça multiplicações entre matrizes, utilizar @ ao invés do *
a = np.array([[1,2,3],[0,1,3]])
b = np.array([[8,12,3],[3,4,5]])
a@b 
# Repare que agora, a multiplicação falha, pois o número de colunas de a é diferente do número de linhas de b
                                                                                                    
c = np.array([[5,6],[4,3],[1,2]])
print(a@c)
print(a.shape, c.shape) # A multiplicação entre a e c ocorre pois os seus shapes estão de acordo com a lei
                                                                                # de multiplicação de matrizes
```

```python
# Funções de agregação
aleatorios = np.random.rand(5,4)

# Soma
print(aleatorios.sum())

# Maior valor (axis=0 ou 1 aplica a função por linha ou coluna )
print(aleatorios.max(axis=0))

# Menor valor
print(aleatorios.min())

# Média
print(aleatorios.mean())

# Desvio Padrão
print(aleatorios.std(axis=1))
```
